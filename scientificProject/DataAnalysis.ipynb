```python
import pandas as pd 
import numpy as np
import plotly.express as px

df = pd.read_csv("Air_Quality.csv")

df
```

```python
fig = px.pie(
    df,
    values="Data_Value",
    names="Name",
    title="Type of air pollutants in the air",
)

fig.show()
```

```python
fig1 = px.density_heatmap(
    df,
    x='Name',
    y='Geo_Type_Name',
    z='Data_Value',
    width=1000,
    height=800,
    labels={'Name':'Nome','Geo_Type_Name':'Geography type name','Data_Value':'Value'}
)

fig1.show()
```

```python
# Estrai anno

df['Start_Date'] = pd.to_datetime(df['Start_Date'], errors='coerce', dayfirst=True)
df['Year'] = df['Start_Date'].dt.year


# Calcola la media di Data_Value per anno e localitÃ 
df_yearly = df.groupby(['Year', 'Geo_Place_Name'])['Data_Value'].mean().reset_index()

# Grafico andamento medio annuale NO2 per localitÃ 
fig2 = px.line(
    df_yearly,
    x='Year',
    y='Data_Value',
    color='Geo_Place_Name',
    line_group='Geo_Place_Name',
    title='Andamento medio annuale di NO2 per localitÃ '
)
fig2.show()

```

```python
def convert_excel_to_csv(file_excel, file_csv, sheet_name=0):
    df = pd.read_excel(file_excel, sheet_name=sheet_name)
    df.to_csv(file_csv, index=False)
    




df_prezzi = pd.read_csv("prezzo_beni_alimentari.csv")

df_prezzi_pulito = pd.melt(df_prezzi, id_vars=["ANNI"], 
                  var_name="Prodotto", 
                  value_name="Prezzo_al_kg")


# Converte Prezzo_al_kg in numerico (float), sostituendo errori con NaN
df_prezzi_pulito["Prezzo_al_kg"] = pd.to_numeric(df_prezzi_pulito["Prezzo_al_kg"], errors='coerce')
# Crea una colonna per il size del bubble con trasformazione logaritmica (valori piccoli + visibili)
df_prezzi_pulito["size_bubble"] = df_prezzi_pulito["Prezzo_al_kg"].apply(lambda x: np.log(x + 1) * 10)

# Rimuove i valori nulli
df_prezzi_pulito = df_prezzi_pulito.dropna(subset=["Prezzo_al_kg"])

fig = px.line(
    df_prezzi_pulito,
    x='ANNI',
    y='Prezzo_al_kg',
    color="Prodotto"
)

fig.show()

```
