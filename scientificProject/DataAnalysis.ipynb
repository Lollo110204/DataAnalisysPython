```python
import pandas as pd 
import numpy as np
import plotly.express as px

df = pd.read_csv("Air_Quality.csv")

display(df)
```

```python
fig = px.pie(
    df,
    values="Data_Value",
    names="Name",
    title="Type of air pollutants in the air",
)

fig.show()
```

```python
fig1 = px.density_heatmap(
    df,
    x='Name',
    y='Geo_Type_Name',
    z='Data_Value',
    width=1000,
    height=800,
    labels={'Name':'Nome','Geo_Type_Name':'Geography type name','Data_Value':'Value'}
)

fig1.show()
```

```python
# Estrai anno

df['Start_Date'] = pd.to_datetime(df['Start_Date'], errors='coerce', dayfirst=True)
df['Year'] = df['Start_Date'].dt.year


# Calcola la media di Data_Value per anno e localitÃ 
df_yearly = df.groupby(['Year', 'Geo_Place_Name'])['Data_Value'].mean().reset_index()

# Grafico andamento medio annuale NO2 per localitÃ 
fig2 = px.line(
    df_yearly,
    x='Year',
    y='Data_Value',
    color='Geo_Place_Name',
    line_group='Geo_Place_Name',
    title='Andamento medio annuale di NO2 per localitÃ '
)
fig2.show()

```

Creiamo una funzione che permetta di fare la conversione da xlxs a csv ( serve nel caso i dati siano su una tabella excell ma a noi servono in csv per poterli anlizxzare meglio)

```python
def convert_to_csv(file_excell,file_csv):
    df = pd.read_excel(file_excell)
    return df.to_csv(file_csv, index=False)



convert_to_csv("WorldBank.xlsx","WorldBank.csv")
```

ora creiamo il df e rinominiamo i campi del nostro csv siccome avere spazi all'interno dei nomi delle colonne danno molti problemi in fase di analisi 

```python
df_world_bank = pd.read_csv("WorldBank.csv")

df_world_bank = df_world_bank.rename(columns={
    'Country Name' : 'Country_name',
    'Country Code' : 'Country_code',
    'Birth rate, crude (per 1,000 people)' : 'Birth_rate_crude',
    'Death rate, crude (per 1,000 people)' : 'Death_rate_crude',
    'Electric power consumption (kWh per capita)'  : 'Electric_power_consumption_kWh_per_capita',
    'GDP (USD)' : 'GDP_USD',
    'GDP per capita (USD)' : 'GDP_per_capita_USD',
    'Individuals using the Internet (% of population)' : 'Individuals_using_Internet_percent_population',
    'Infant mortality rate (per 1,000 live births)' : 'Infant_mortality_rate_per_1000_live_births',
    'Life expectancy at birth (years)' : 'Life_expectancy_at_birth_years',
    'Population density (people per sq. km of land area)' : 'Population_density_people_per_sq_km_land_area',
    'Unemployment (% of total labor force) (modeled ILO estimate)' : 'Unemployment_percent_total_labor_force_modeled_ILO_estimate'
})

display(df_world_bank)
display(df_world_bank.describe())
```

```python


fig = px.bar(
    df_world_bank,
    x="Country_name",
    y="Death_rate_crude",
    color="Country_name",
    labels={"Country_name": "Nome del paese", "Death_rate_crude": "Tasso di mortalitÃ  per anno (per 1000 abitanti)"},
    title="Tasso di mortalitÃ  per paese"
)

fig.show()
```
